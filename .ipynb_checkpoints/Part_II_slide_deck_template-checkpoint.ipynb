```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
bikes_clean = pd.read_csv('bikes_clean.csv')
```

```python
def myplot(df, xvar, hueVar=None,color=0, order=None ):
  #plot dimensions
  plt.figure(figsize=(10,7))
  #plot
  sns.countplot(data=df,x=xvar,hue=hueVar,color=sns.color_palette()[0],order=order,edgecolor='black');
  xvar=xvar.replace('_'," ")
  if hueVar:
    hueVar=hueVar.replace("_", " ")
  plt.title(f"Distribution of {xvar} {'by' if hueVar else ''} {hueVar if hueVar else ''} ".title(), fontsize=14, weight = 'bold' )
  plt.xlabel(xvar.title(),fontsize = 10, weight = 'bold' )
  plt.ylabel('Frequency'.title(), fontsize=10,weight='bold')
```

## Percentage of User types

> The Pie chart illustrates the percentages of both user types. There are 183,412 rows contained in the bike share dataset and we find there are a large percentage of subscribers in the program to customers who are not enrolled in the program.


```python
sorted_counts = bikes_clean['user_type'].value_counts()
 #show bike user type percentage
plt.figure(figsize=(7,7))
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,
        counterclock = False, autopct='%1.2f%%');

plt.title('Bike User Count Comparison', fontsize=15);
```

## Usage of Service based on day of the week

>This service is used more on weekdays than weekends

```python
day_name = ["Sunday", "Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]

#Show daily bike use

myplot(bikes_clean,'day',order=day_name)
```

## Bike Service Hours for both users

>Most Subscribers use the service at peak hours between 8 and 17 (5pm).

>Customers have more flexible time, mostly used in the afternoon.





```python
myplot(bikes_clean,'hour_of_day',hueVar='user_type')
```

## Bike service usage per day for both users
